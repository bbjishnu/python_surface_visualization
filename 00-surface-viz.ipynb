## 2. Surface Plot  of $ f(x,y) =\sqrt{x^2+y^2} $


```python
def mfunc(x,y):
    return np.sqrt(x**2+y**2)    
```

```python
# Simulate Data 
theta = np.linspace(0,2*np.pi,100)
radi = np.linspace(0, 10,50)
T, R = np.meshgrid(theta,radi)
# Then calculate X, Y, and Z
X = R * np.cos(T)
Y = R * np.sin(T)
Z = np.sqrt(X**2 + Y**2) 
```

#### 2.1 Surface Plot 

```python

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='winter',linewidth=0, antialiased=False)
# ax.set_axis_off()
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
```

#### 2.2 Wireframe Plot 

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, Z, color="orangered")
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
```

#### 2.3 3D Contour Plot

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.contour3D(X, Y, Z, 25, cmap='jet')
# ax.set_axis_off()
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
```

```python
ax = plt.axes(projection='3d')
X = np.reshape(X,-1)
Y = np.reshape(Y,-1)
Z = np.reshape(Z,-1)
# ax.plot_trisurf(X, Y, Z, cmap='hsv')
ax.plot_trisurf(X, Y, Z, cmap='Set1')
# ax.set_axis_off()
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
```

```python
ax = plt.axes(projection='3d')
X = np.reshape(X,-1)
Y = np.reshape(Y,-1)
Z = np.reshape(Z,-1)
# ax.plot_trisurf(X, Y, Z, cmap='hsv')
ax.scatter(X, Y, Z, c=Z**2, cmap='PuBuGn_r', linewidth=10)
# ax.set_axis_off()
ax.grid(False)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
```
